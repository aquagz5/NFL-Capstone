```python
subset = data[['gameId', 'playId', 'nflId', 'frameId', 'tackle_binary_all', 'tack_prob']]
subset = subset.drop_duplicates(subset = ['gameId', 'playId', 'nflId', 'frameId'], keep = 'first').reset_index(drop = True) 
# subset['frame_num'] = subset.groupby(['gamePlayId', 'nflId'])['frameId'].rank(method='first')
subset
```

```python
def player_eval_play(df, prob_col, start_frame=None, end_frame=None):
    '''
    This function takes in the dataframe and shows which players gained/lost the most tackle probability over a given play segment (defaults to entire play).
    '''
    if start_frame is not None and end_frame is not None:
        if start_frame>=end_frame:
            raise ValueError('start_frame must be less than end_frame')
        
    df['frame_num'] = df.groupby(['gameId', 'playId', 'nflId'])['frameId'].rank(method='first')

    min_df = df.groupby(['gameId', 'playId', 'nflId'])['frame_num'].min().reset_index()
    max_df = df.groupby(['gameId', 'playId', 'nflId'])['frame_num'].max().reset_index()
    
    min_df['start_frame'] = min_df['frame_num'] if start_frame is None else start_frame
    max_df['end_frame'] = max_df['frame_num'] if end_frame is None else end_frame
    
    min_df['use_frame'] = min_df[['start_frame', 'frame_num']].max(axis=1)
    max_df['use_frame'] = max_df[['end_frame', 'frame_num']].max(axis=1)
    
    lower = df.merge(min_df[['gameId', 'playId', 'nflId', 'use_frame']], left_on=['gameId', 'playId', 'nflId', 'frame_num'], right_on=['gameId', 'playId', 'nflId', 'use_frame'], how='inner')
    upper = df.merge(max_df[['gameId', 'playId', 'nflId', 'use_frame']], left_on=['gameId', 'playId', 'nflId', 'frame_num'], right_on=['gameId', 'playId', 'nflId', 'use_frame'], how='inner')    

    final = lower[['gameId', 'playId', 'nflId', prob_col]].merge(upper[['gameId', 'playId', 'nflId', prob_col]], on=['gameId', 'playId', 'nflId'], how='inner')
    
    final['diff'] = final[prob_col + '_x'] - final[prob_col + '_y']
    
    output = final[['gameId', 'playId', 'nflId', 'diff']]
    
    return output
```

```python
player_eval_play(subset, 'tack_prob')
```
